# Detection and Correction of Outliers 

```python
def outlier(df):
    outliers_treated = list()
    ContinuousVars=financeD.columns[(financeD.dtypes!='O')].drop(['Revenue.Grid','year_last_moved'])
    for column in ContinuousVars:
        #print(column)
    
        Q1=np.percentile(financeD[column],25)
        Q3=np.percentile(financeD[column],75)
        IQR=Q3-Q1
    
        UB=Q3+1.5*IQR
        LB=Q1-1.5*IQR
    
    
    ########## Correction of outliers
    
    
        if (min(df[column]) < LB) or (max(df[column]) > UB):
            outliers_treated.append(column)
    
    ## UPPER BOUND operations
        if min(financeD[column])>UB:
            financeD[column]=np.where(financeD[column]>UB,UB,financeD[column])
            #outliers_treated.append(column)
    
    
    ## LOWER BOUND operations
        if min(financeD[column])<LB:
            financeD[column]=np.where(financeD[column]<LB,LB,financeD[column])
            #outliers_treated.append(column)
            
            
    print("Outliers Treated:-",outliers_treated)
```

```python
def label_encoder(col,start_point=0):
    value_df=financeD[col].value_counts().sort_values()
    for i in range(len(value_df)):
        value_to_encode=value_df.index[i]
        financeD[col][financeD[col]==value_to_encode]=i+start_point
    financeD[col]=financeD[col].astype(np.int16)
```
